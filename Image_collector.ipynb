```python
# Import opencv
import cv2 

# Import uuid
import uuid

# Import Operating System
import os

# Import time
import time
```

```python
paths = {
'VIDEO_PATH' : os.path.join(os.path.join(os.getcwd()),'video_dataset'),
'RAW_DATA_PATH' : os.path.join(os.path.join(os.getcwd()),'raw_images'), 
'RAW_DATA_PATH_SPLITTED' : os.path.join(os.path.join(os.getcwd()),'raw_images_splitted'), 
'DATA_PATH' : os.path.join(os.path.join(os.getcwd()), 'train_data'),
'DATA_PATH_IMAGES' : os.path.join(os.path.join(os.getcwd()), 'train_data', 'images'),
'DATA_PATH_LABELS' : os.path.join(os.path.join(os.getcwd()), 'train_data', 'labels'),
'DATA_PATH_IMAGES_TRAIN' : os.path.join(os.path.join(os.getcwd()), 'train_data', 'images', 'train'),
'DATA_PATH_IMAGES_VAL' : os.path.join(os.path.join(os.getcwd()), 'train_data', 'images', 'val'),
'DATA_PATH_LABELS_TRAIN' : os.path.join(os.path.join(os.getcwd()), 'train_data', 'labels', 'train'),
'DATA_PATH_LABELS_VAL' : os.path.join(os.path.join(os.getcwd()), 'train_data', 'labels', 'val'),
}
```

```python
for path in paths.values():
    if not os.path.exists(path):
        os.mkdir(path)
```

```python
files = os.listdir(paths['VIDEO_PATH'])
video_list = []
for file in files:
    if os.path.isfile(os.path.join(paths['VIDEO_PATH'], file)):
        full_path = os.path.join(paths['VIDEO_PATH'], file)
        video_list.append(full_path)
print(video_list)
```

```python
for vid in video_list:
    cam = cv2.VideoCapture(vid)
    basename = os.path.basename(vid)
    label = os.path.splitext(basename)[0]
    path = os.path.join(paths['RAW_DATA_PATH'], label)
    print(path)
    try:
        if not os.path.exists(path):
            os.makedirs(path)

    except OSError:
        print ('Error: Creating directory of data')
        

    # frame
    currentframe = 0

    while(True):
        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = path + '/'+label+'_frame' + str(currentframe) + '.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1
        else:
            break

    # Release all space and windows once done
    cam.release()
    cv2.destroyAllWindows()
```
